***Implementar um algoritmo de conversão de um numero na base decimal para seu respectivo correspondente na base Binária.***
 

```python
numero_base_decimal = 23
numero_base_binaria = "{0:b}".format(numero_base_decimal)
print(f'Numero na base Decimal: {numero_base_decimal}')
print(f'Numero na base Binária: {numero_base_binaria}')
```

```python
numero_base_decimal = int(input('Digite um numero de base decimal inteiro para vê-lo em base binária: '))
numero_base_binaria = "{0:b}".format(numero_base_decimal)
print(f'Numero na base Decimal: {numero_base_decimal}')
print(f'Numero na base Binária: {numero_base_binaria}')
```
